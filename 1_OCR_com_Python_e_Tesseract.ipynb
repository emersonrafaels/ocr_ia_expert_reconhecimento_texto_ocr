## Realizando a leitura da imagem

```python
img = cv2.imread('../Imagens/teste_manuscrito_01.jpg')
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(rgb)
plt.show()
```

*texto em itálico*- block_num = Número do bloco atual. Quando o tesseract faz o OCR, ele divide a imagem em várias regiões, o que pode variar de acordo com os parametros do PSM e também outros critérios próprios do algoritmo. Cada bloco é uma região

- conf = confiança da predição (de 0 a 100. -1 significa que não foi reconhecido texto)

- height = altura do bloco de texto detectada (ou seja, da caixa delimitadora)

- left = coordenada x onde inicia a caixa delimitadora

- level = o level (nível) corresponde à categoria do bloco detectado. são 5 valores possiveis:
  1. página
  2. bloco
  3. parágrafo
  4. linha
  5. palavra

Portanto, se foi retornado o valor 5 significa que o bloco detectado é texto, se foi 4 significa que o que foi detectado é uma linha

- line_num = número da linha do que foi detectado (inicia com 0)

- page_num = o índice da página onde o item foi detectado. Na maioria dos casos sempre haverá uma página só

- text = o resultado do reconhecimento

- top = coordenada y onde a caixa delimitadora começa

- width = largura do bloco de texto atual detectado

- word_num = numero da palavra (indice) dentro do bloco atual

## Realizando o image to data

```python
resultado = pytesseract.image_to_data(rgb, lang='por', output_type=Output.DICT)
resultado
```

```python
min_conf = 20 #@param {type: 'slider', min: 0, max: 100}
```

```python
def caixa_texto(resultado, img, cor=(255, 100, 0)):

  # OBTENDO OS PARÂMETROS:
  # X0 = LEFT, YO = TOP
  # XF = LEFT + WIDTH, YF = TOP + HEIGHT

  x = resultado['left'][i]
  y = resultado['top'][i]
  w = resultado['width'][i]
  h = resultado['height'][i]

  # OS PARÂMETROS DO CV2.RECTANGLE SÃO:
  # IMAGEM
  # COORDENADAS
  # COR DO RETÂNGULO = cor = (255, 100, 0
  # ESPESSURA DO RETÂNGULO = 2
  cv2.rectangle(img, (x, y), (x + w, y + h), cor, 2)

  return x, y, img
```

```python
# QUANTIDADE DE TEXTO INICIAL
# MUITOS SENDO TEXTO '-'

print("QUANTIDADE DE TEXTO: {}".format(len(resultado['text'])))
resultado['text']
```

```python
img_copia = rgb.copy()

# PERCORRENDO CADA UM DOS TEXTOS OBTIDOS
for i in range(len(resultado['text'])):

  # OBTENDO O INTERVALO DE CONFIANÇA
  confianca = int(resultado['conf'][i])

  # REALIZANDO O THRESHOLD DE CONFIANÇA
  if confianca > min_conf:

    # REALIZANDO A CRIAÇÃO DA CAIXA DE TEXTO SOBRE O TEXTO
    x, y, img = caixa_texto(resultado, img_copia)
    
    # OBTENDO O TEXTO ATUAL
    texto = resultado['text'][i]

    # INSERINDO O TEXTO SOBRE A CAIXA RETANGULAR (TOP - 10)
    cv2.putText(img_copia, texto, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 1.1, (0, 0, 255))

plt.imshow(img_copia)
plt.show()
```

### Escrevendo o texto sobre a imagem

```python
def escreve_texto(texto, x, y, img, fonte, tamanho_texto=32, cor=(0, 0, 255)):

    try:
        fonte = ImageFont.truetype(fonte, tamanho_texto)

        img_pil = Image.fromarray(img)

        draw = ImageDraw.Draw(img_pil)

        draw.text((x, y - tamanho_texto), texto, font=fonte, fill=cor)

        img = np.array(img_pil)

    except Exception as ex:
        print(ex)

    return img
```

```python
img_copia = rgb.copy()

# PERCORRENDO CADA UM DOS TEXTOS OBTIDOS
for i in range(len(resultado['text'])):

    # OBTENDO O INTERVALO DE CONFIANÇA
    confianca = int(resultado['conf'][i])

    # REALIZANDO O THRESHOLD DE CONFIANÇA
    if confianca > min_conf:

        # REALIZANDO A CRIAÇÃO DA CAIXA DE TEXTO SOBRE O TEXTO
        x, y, img = caixa_texto(resultado, img_copia)

        # OBTENDO O TEXTO ATUAL
        texto = resultado['text'][i]

        # INSERINDO O TEXTO SOBRE A CAIXA RETANGULAR (TOP - 10)
        img_copia = escreve_texto(texto, x, y, img_copia, fonte)

plt.imshow(img_copia)
plt.show()
```

## Definindo um pattern a ser buscado

```python
padrao_data = '^(0[1-9]|[12][0-9]|3[01])/(0[1-9]|1[012])/(19|20)\d\d$'
```

## Buscando o pattern

```python
# INICIANDO A LISTA QUE ARMAZENARÁ TODAS AS DATAS
datas = []

img_copia = rgb.copy()

# PERCORRENDO CADA UM DOS TEXTOS OBTIDOS
for i in range(0, len(resultado['text'])):

  # OBTENDO O INTERVALO DE CONFIANÇA
  confianca = int(resultado['conf'][i])

  # REALIZANDO O THRESHOLD DE CONFIANÇA
  if confianca > min_conf:

    # OBTENDO O TEXTO ATUAL
    texto = resultado['text'][i]

    # VERIFICANDO SE O TEXTO POSSUI O PADRÃO DE REGEX BUSCADO
    if re.match(padrao_data, texto):

      # CASO POSSUA O PADRÃO DE REGEX COLOCA O TEXTO
      # INSERINDO O TEXTO SOBRE A CAIXA RETANGULAR (TOP - 10)
      img_copia = escreve_texto(texto, x, y, img_copia, fonte, 12)
      datas.append(texto)
    
    # CRIA O BOUNDING BOX SOBRE O TEXTO
    x, y, img_copia = caixa_texto(resultado, img_copia)


plt.imshow(img_copia)
plt.show()
```

## Retirando os falsos positivos

```python
img_copia = rgb.copy()

# PERCORRENDO CADA UM DOS TEXTOS OBTIDOS
for i in range(len(resultado['text'])):

  # OBTENDO O INTERVALO DE CONFIANÇA
  confianca = int(resultado['conf'][i])
  text = resultado['text'][i]

  # REALIZANDO O THRESHOLD DE CONFIANÇA
  if confianca > min_conf and not text.isspace() and len(text) > 1:

    # REALIZANDO A CRIAÇÃO DA CAIXA DE TEXTO SOBRE O TEXTO
    x, y, img = caixa_texto(resultado, img_copia)
    
    # OBTENDO O TEXTO ATUAL
    texto = resultado['text'][i]

    # INSERINDO O TEXTO SOBRE A CAIXA RETANGULAR (TOP - 10)
    img_copia = escreve_texto(texto, x, y, img_copia, fonte)

plt.imshow(img_copia)
plt.show()
```
