# Inversão

```python
img = cv2.imread('../Imagens/img-process.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap="gray")
plt.show()

texto = pytesseract.image_to_string(gray)
print(texto)
```

```python
# VISUALIZANDO A MATRIZ ORIGINAL (FUNDO ESCURO E LETRAS CLARAS)

gray
```

```python
# REALIZANDO O PROCESOS DE INVERSÃO, SUBTRAIMOS O VALOR MAIS CLARO (255) (FUNDO CLARO E LETRAS ESCURAS)

invert = 255 - gray
```

```python
invert
```

```python
plt.imshow(invert, cmap="gray")
plt.show()

texto = pytesseract.image_to_string(invert)
print(texto)
```

```python
# APLICANDO UM LIMIAR DE OTSU

val, invert_thesh_otsu = cv2.threshold(invert, thresh=0, maxval=255, type=cv2.THRESH_BINARY | cv2.THRESH_OTSU)
```

```python
plt.imshow(invert_thesh_otsu, cmap="gray")
plt.show()

texto = pytesseract.image_to_string(invert_thesh_otsu)
print(texto)
```

```python
invert_invert_thesh_otsu = 255 - invert_thesh_otsu
plt.imshow(invert_invert_thesh_otsu, cmap="gray")
plt.show()

texto = pytesseract.image_to_string(invert_invert_thesh_otsu)
print(texto)
```

# Redimensionamento

- INTER_NEAREST - uma interpolação de vizinho mais próximo. É muito usado por ser o mais rápido.
- INTER_LINEAR - uma interpolação bilinear (é usada por padrão), boa no geral para aumentar e também pra diminuir imagens.
- INTER_AREA - usa a relação de área de pixel. Pode ser um método preferido para a redução de imagens pois fornece resultados sem moiré (efeito geralmente indesejado na imagem). Mas quando a imagem é ampliada, é semelhante ao método INTER_NEAREST.
- INTER_CUBIC - bicúbica (4x4 pixel vizinhos). Possui resultados melhores.
- INTER_LANCZOS4 - interpolação Lanczos (8x8 pixel vizinhos). Dentre esses algoritmos, é o que apresenta resultados com a melhor qualidade.

```python
plt.imshow(gray, cmap="gray")
plt.show()
```

```python
# NESSE CASO, ESTAMOS USANDO A INTERPOLAÇÃO INTER CUBIC, MUITO USADA PARA 4X4 PIXELS VIZINHOS SEREM INTERPOLADOS.

maior = cv2.resize(gray, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_CUBIC)
plt.imshow(maior, cmap="gray")
plt.show()
```

```python
# NESSE CASO, ESTAMOS A INTERPOLAÇÃO INTER AREA, POIS É RÁPIDA E É O METODO PREFERÍVEL PARA REDUÇÃO DE IMAGENS

menor = cv2.resize(gray, None, fx = 0.5, fy = 0.5, interpolation=cv2.INTER_AREA)
cv2_imshow(menor)
```
